## C. Les listes

Comme on l'a vu, une **liste** est une **séquence ordonnée d'éléments modifiables de types variables**. On accède à un de ses éléments grâce à son **indice entre crochets**.

### I. Construire des listes

#### I.1 La base

Une liste s'écrit entre <span style="color:red;">**crochets**</span> et ses éléments sont séparés par des <span style="color:red;">**virgules**</span>.

```python
#Exemple 1 :
L = [4, 1, 2, 6, 7] # création d'une liste de 5 éléments
print(type(L)) # affichage du type
print(L) #affichage de la liste
print(L[1]) #affichage du 2ème élément
print(L[3]) #affichage du 4ème élément
print(L[-1]) #affichage du dernier élément (premier en partant de la fin)
print(L[1]+L[2]) #affichage du résultat de la somme des 2ème et 3ème élément
```

#### I.2 Construire une liste à partir d'une autre liste par agrandissement
Une **liste vide** s'écrit `[]`. Comme pour les tuples, l'opérateur `+` permet de **concaténer** des listes, et l'opérateur `*` permet de **copier plusieurs fois à la suite la même liste**.

De plus, on peut **ajouter un élément à la fin d'une liste** (vide ou non) grâce à la méthode `.append()` et la méthode `.extend()` permet **d'ajouter plusieurs éléments d'un coup à la fin d'une liste** (concaténation).

**Remarque** : Les méthodes `.append()` et `.extend()` ne peuvent pas être utilisées pour des tuples puisqu'ils ne sont pas modifiables.

```python

#Exemple 2 :
L1=[] #L1 est une liste vide
print(L1)
L2=[4] #L2 est une liste d'un seul élement
print(5*L2) #on affiche 5 copies de L2 à la suite
print(L2)
L2=L1+L2 #L1+L2=L2 puisque L1 est vide
print(L2) 
L2.append(5) #on ajoute un élément à la fin de L2
print(L2)
L2.extend([6,7,8]) #on modifie L2 en y ajoutant 3 éléments
print(L2)
L2=L2+[9,10] #on ajoute encore 2 nouveaux éléments à la suite de L2
print(L2)
```

#### I.3 Construire une liste par conversion
La fonction `list()` permet de **convertir** (quand c'est possible) un objet en liste.

La fonction `len()` renvoie le **nombre d'éléments** dans une liste.

```python
#Exemple 3 :
chaine="J'aime toujours faire du python" #chaîne de caractères
L=list(chaine) #conversion en liste, on le peut car une chaîne de caractère est itérable
print(L)
print(len(chaine), len(L)) #la chaîne de caractères et la liste ont la même longueur
print(list(range(5))) #range(5) renvoie les valeurs de 0 à 4
print(list(4)) #message d'erreur car un entier n'est pas itérable, on ne peut pas le convertir en liste
```

#### I.4 Construire une liste par itération

Lorsque l'on souhaite contruire une **liste dont les éléments suivent une certaine logique** (des nombres pairs par exemple), on peut utiliser une boucle `for` et ajouter les éléments un par un grâce à `.append`.

Mais Python permet aussi de le faire très facilement avec les instructions suivantes :
- `[f(x) for x in iterable]`
- `[f(x) for x in iterable if test]`.

```python
#Exemple 4 :
print([x**2 for x in range(10)]) #liste contenant le carré des 10 premiers entiers
print([x//2 for x in [3,1,7]]) #liste contenant le reste de la division entière par 2 des nombres 3, 1 et 7
print( [x for x in range(50) if x%3==0 and x%4!=0]) # liste des entiers inférieurs à 50 qui sont des multiples de 3 mais pas de 4
```

### II. Manipuler des listes

On a déjà vu qu'on pouvait concaténer des listes avec `+` et accèder à la valeur d'un élément seul avec `ma_liste[i]`. On a vu aussi les méthodes `.append()` et `.extend()` qui permettent de modifier une liste. Voyons ce qu'on peut faire de plus ...

#### II.1 Extraire un morceau de liste (slice)

On peut, comme pour les tuples, **extraire des morceaux de liste** (slice) avec `ma_liste[i:j]` qui renvoie une liste composée des éléments d'indices **$i$ à $j-1$**. Si **$i$ n'est pas précisé**, on extrait **à partir du premier** élément et si **$j$ n'est précisé**, on extrait **jusqu'au dernier**.

```python
#Exemple 5:
L=[4,1,5,2,3,9,7,2,8]
print(L[2:5]) #affiche du 3ème(indice2) au 5ème(indice 4) élément
print(L[:5]) #affiche du 1er(indice 0) au 5ème(indice 4) élément
print(L[3:]) #affiche du 4ème(indice 3) au dernier élément
L2=L[:3]+[10]+L[4:] #crée une nouvelle liste dont le 4ème élément vaut 10 au lieu de 2
print(L2) #cette nouvealle liste est une copie modifiée de la première
L[3]=10 #mais contrairement aux tuples, on peut aussi directement modifier la liste de départ
print(L)
```

#### II.2 Supprimer un élément d'une liste
On peut aussi **supprimer un élément** d'une liste **à partir de son indice** grâce à l'instruction `del` ou **à partir de sa valeur** grâce à la méthode `.remove()`.

La méthode `.pop(i)` permet à la fois **d'enlever l'élément d'indice $i$ tout en retournant sa valeur**. Si $i$ n'est pas précisé, c'est le dernier élément de la liste qui est "poppé".

```python
#Exemple 5 :
L=[4, 1, 2, 6, 7, 15, 7]
del L[2] #supprime le troisième élément de L
print(L)
L.remove(15) #supprime le premier élément de L qui vaut 15
print(L)
x=L.pop(2) #affecte à x la 3ème valeur de L et l'enlève de L
print(x)
print(L)
```

#### II.3 Insérer un élément dans une liste
On peut  **insérer un élément à une position donnée** dans une liste grâce à la méthode `.insert(i,x)` qui insère la valeur $x$ à l'indice $i$.

**Remarque :** `.insert(0,x)` permet d'insérer $x$ au début de la liste et `.insert(-1,x)` d'insérer $x$ à la fin de la liste.

```python
#Exemple 6:
L=[ 3 , 6 , 2 , 7 , 1 , 9 , 5]
L.insert(2,0) #insert 0 en 3ème position
print(L)
```

#### II.4 Autres méthodes pour modifier une liste

La méthode `liste.index(x)` renvoie **l'indice du premier élément trouvé** de la liste **dont la valeur vaut $x$**.

La méthode `liste.count(x)` renvoie le **nombre d'éléments trouvés** dans la liste **dont la valeur vaut $x$**.

La méthode `liste.sort()` **modifie la liste** pour qu'elle soit **triée dans l'odre croissant** (*tri sur place*) alors que la fonction `sorted(liste)` **renvoie la liste triée dans l'ordre croissant**.

La méthode `liste.reverse()` **modifie la liste en inversant l'ordre de ses éléments**.

```python
#Exemple 7:
L=[ 3 , 6 , 2 , 7 , 1 , 9 , 5, 6]
print(L.index(6)) #indice du premier élément qui vaut 6
print(L.count(6)) #nombre d'éléments qui valent 6
L2=sorted(L) #crée une liste L2 qui est la version triée de L
print(L2)
L.sort() #trie la liste L sur place
print(L)
L.reverse() #inverse l'ordre des éléments de L
print(L)
```

#### II.5 Copier une liste

Si `L1` est une liste, lorsqu'on écrit `L2=L1`, on ne **crée pas une copie de `L1` qui s'appelerait `L2`**. On donne juste **deux noms différents à une seule et même liste**. Ceci est une source fréquente de "bugs".

Pour faire une **copie indépendante** d'une liste, on peut utiliser la fonction `list()`.

```python
#Exemple 8:
L1=[2,4,1,6,5,7]
L2=L1
L1.append(3) #on ajoute 3 à la fin de L1
print("L1 :", L1)
print("L2 :", L2) # on s'apercoit que L2 reste identique à L1 après modification
L2=list(L1) #on crée une vraie copie de L1
L1.append(3) #on ajoute à nouveau 3 à la fin de L1
print("L1 :", L1)
print("L2 :", L2) #cette fois L2 n'a pas été modifiée en même temps que L1
```
