## 2. Faire la même chose en Python

Le script suivant permet de lire et de créer une liste de listes comprenant toutes les données du fichier csv.

````python
# chargement de la bibliothèque csv
import csv
# chargement du fichier csv
with open("Files/lycee.csv","r", encoding="utf-8") as f:
    premiere_ligne = f.readline()# stockage de la première ligne (descripteurs)
    # affichage de ces desripteurs
    print("Les descripteurs de la collection des établissements scolaires :")
    print(premiere_ligne)
    # Ces deux lignes stockent les éléments du fichier dans une liste (data_lignes) # en omettant la première ligne (sniffer)
    dialecte_fichier_csv = csv.Sniffer().sniff(premiere_ligne)
    data_lignes = list(csv.reader(f, dialect=dialecte_fichier_csv))
# on affiche ici la première ligne de la liste data_lignes
print(data_lignes[0])
````

**Q2. Écrire un code qui demande à l’utilisateur un nom de ville et qui affiche les adresses de tous les établissements scolaires de cette ville**

````python
# Question 2
# Pour ne pas avoir à compter dans quelle colonne se trouve chaque descripteur, on peut utilise le dictionnaire suivant
premiere_ligne = list(premiere_ligne.rstrip().split(",")) #convertit la 1ere ligne en liste après l'avoir découpée
dico_colonnes={premiere_ligne[i]:i for i in range(len(premiere_ligne))} # les clés sont les noms des descripteurs et les valeurs sont les numéros de colonne correspondant
col_ville=dico_colonnes.get('Commune') #col_ville contient le numéro de colonne où se trouve la commune

ville=input("Dans quelle ville cherchez-vous un établissement scolaire ? ")
for ligne in data_lignes :
    if ligne[col_ville]==ville :
        print("Nom :", ligne[dico_colonnes.get('Appellation officielle')])
        print("Adresse :", ligne[dico_colonnes.get('Adresse')])
        print("Code postal :", ligne[dico_colonnes.get('Code postal')])
        print("Ville :", ligne[col_ville],"\n")
````
