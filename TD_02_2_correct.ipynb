**Q1. Écrire une fonction qui prend en paramètre une liste d’entiers naturels et qui renvoie la somme des éléments de la liste.**

```python
#Question 1 :
def somme(tab):
    for nb in tab:
        assert type(nb)== int and nb>0, "Les valeurs de la liste doivent être des entiers naturels."
    result=0
    for nb in tab : #nb prend les valeurs successives des éléments de tab
        result=result+nb #à chaque tour de boucle, on ajoute une nouvelle valeur
    return result
L=[1,2,3,4,5]
print(somme(L))
```

**Q2. Écrire une fonction qui prend en paramètre une liste d’entiers naturels et qui renvoie le plus grand des éléments de la liste.**

```python
#Question 2 :
def maximum(tab):
    for nb in tab:
        assert type(nb)== int and nb>0, "Les valeurs de la liste doivent être des entiers naturels."
    result=0 #on initialise le résultat à 0
    for nb in tab: #nb parcourt les valeurs de tab
        if nb>result : result=nb #si nb est plus grand que result, il devient le nouveau maximum
    return result
L=[3,1,7,13,4,9]
print(maximum(L))
```

**Q3. Écrire une fonction qui prend en paramètre une liste d’entiers naturels et qui renvoie le plus petit des éléments de la liste.**

```python
#Question 3 :
def minimum(tab):
    for nb in tab:
        assert type(nb)== int and nb>0, "Les valeurs de la liste doivent être des entiers naturels."
    result=tab[0] #on initialise result avec la première valeur de tab
    for nb in tab:
        if nb<result : result=nb #si nb est plus petit que result, il devient le nouveau minimum
    return result
L=[8,13,7,13,4,9]
print(minimum(L))
```

**Q4. Écrire une fonction qui prend en paramètre un entier naturel et qui renvoie la liste des nombres pairs inférieurs à cet entier.**

```python
#Question 4 :
def pairs(n):
    assert type(n) == int and n>0, "L'argument doit être un entier naturel."
    result=[] #on initialise result avec une liste vide
    for i in range((n+1)//2): #i<(n+1)//2 => 2*i<n si n pair et 2*i<n+1 soit 2*i<=n-1 si n impair
        result.append(2*i) #on ajoute les nombres pairs sans dépasser n
    return result
print(pairs(16))
```

**Q5. Écrire une fonction qui prend en paramètre un entier naturel et qui renvoie la liste des nombres impairs inférieurs à cet entier.**

```python
#Question 5 :
def impairs(n):
    assert type(n) == int and n>0, "L'argument doit être un entier naturel."
    result=[]
    for i in range(n//2): #i<n//2 => 2*i+1<n si n impair et 2*i+1<n+1 soit 2*i+1<=n-1 si n est pair
        result.append(2*i+1)
    return result
print(impairs(18))
```

**Q6. Écrire une fonction qui prend en paramètre un entier naturel et qui renvoie la liste des multiples de 7 inférieurs à cet entier.**

```python
#Question 6 :
def multiples_de_sept(n):
    assert type(n) == int and n>0, "L'argument doit être un entier naturel."
    result=[]
    for i in range(1,n): #on teste les i de 1 à n-1 juste pour éviter de renvoyer 0 mais on peut aussu démarrer de 7
        if i%7 == 0 : result.append(i) # si le reste de la division entière de i par 7 est nul alors i est un multiple de 7
    return result
print(multiples_de_sept(22))
```

**Q9. Écrire une fonction qui supprime les éléments en double d’une liste d'entiers.**

Exemple d’exécution :

liste de départ : `L=[1,2,5,8,6,2,5,9,1,8,8]`

liste d'arrivée : `L=[1, 2, 5, 8, 6, 9]`

```python
#Question 9 :
def del_doublons(tab):
    for nb in tab:
        assert type(nb)== int , "Les valeurs de la liste doivent être des entiers."
    for i in range(len(tab)-1,1,-1): #on parcourt la liste en partant de la fin
        if tab[i] in tab[:i-1]: del tab[i] #si tab[i] existe déjà dans les valeurs précédentes, on le supprime
L=[1,2,5,8,6,2,5,9,1,8,8]
del_doublons(L) #cette fonction ne retourne rien mais modifie directement la liste donnée en argument
print(L) #on affiche la liste modifiée
```

**Q10. Écrire un programme qui prend en argument deux listes de nombres `val` et `coeff` et qui renvoie la moyenne des valeurs de la liste `val` pondérée par les coefficients de la liste `coeff`.**

Exemple :
Si `val = [12.5, 13.6, 18.4, 9.7]` et `coeff = [2, 3, 5, 4]`, la moyenne pondérée vaut :

$$\frac{12.5\times 2 + 13.6\times 3 + 18.4\times 5 + 9.7\times 4}{2 + 3 + 5 + 4} = 14.042857143$$.

```python
#Question 10 :
def moyenne_ponderee(val, coeff):
    for nb in val:
        assert type(nb)== int or type(nb)==float , "Les valeurs de la liste doivent être des nombres."
    for nb in coeff:
        assert type(nb)== int or type(nb)==float , "Les valeurs des coefficients doivent être des nombres."
    assert len(val)==len(coeff), "Les listes de valeurs et de coefficients doivent avoir la même longueur."
    result=0
    for i in range(len(val)):
        result=result+val[i]*coeff[i]
    return result/somme(coeff) #on réutilise la fonction somme définie à la question 1
Val = [12.5, 13.6, 18.4, 9.7]
Coeff = [2, 3, 5, 4]
print(moyenne_ponderee(Val,Coeff))
```
