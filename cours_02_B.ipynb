## B. Les n-uplets

Comme on l'a vu, un **tuple** (objet de type n-uplet) est une **séquence ordonnée d'éléments non modifiables et de types variables**. On accède à un de ses éléments grâce à son **indice entre crochets**.

### I. Construire des tuples

#### I.1 La base

Un tuple s'écrit entre <span style="color:red;">**parenthèses**</span> et ses éléments sont séparés par des <span style="color:red;">**virgules**</span>.

```python
#Exemple 1 :
mon_tuple=(1,2,3,'bonjour') # création d'un tuple de 4 éléments
print(type(mon_tuple)) # affichage du type
print(mon_tuple) #affichage du tuple
print(mon_tuple[1]) #affichage du 2ème élément
print(mon_tuple[3]) #affichage du 4ème élément
print(mon_tuple[-1]) #affichage du dernier élément (premier en partant de la fin)
print(mon_tuple[1]+mon_tuple[2]) #affichage du résultat de la somme des 2ème et 3ème élément
mon_tuple[2]=2 #renvoie une erreur car un tuple n'est pas mutable (modifiable)
```

#### I.2 Construire un tuple à partir d'un autre tuple par agrandissement

Un **tuple vide** s'écrit `()` et un **tuple  composé d'une seule valeur** (singulet) doit quand même comporter une **virgule après la valeur**.

L'opérateur `+` permet de **concaténer** des tuples et l'opérateur `*` permet de **copier plusieurs fois à la suite le même tuple**.

```python
#Exemple 2 :
t1=() #t1 est un tuple vide
print(t1)
t2=(4,) #t2 est un singulet
print(t2)
print(t1+t2) #t1+t2=t2 puisque t1 est vide
print(t2+(5,6,7)) #on ajoute 3 nouveaux éléments à la suite de t2
print(5*t2) #on copie 5 fois à la suite le tuple t2
```

#### I.3 Construire un tuple par conversion

La fonction `tuple()` permet de **convertir** (quand c'est possible) un objet en tuple.

La fonction `len()` renvoie le **nombre d'éléments** dans un tuple.

```python
#Exemple 3 :
chaine="J'aime faire du python" #chaîne de caractères
t=tuple(chaine) #conversion en tuple, on le peut car une chaîne de caractère est itérable
print(t)
print(len(chaine), len(t)) #la chaîne de caractères et le tuple ont la même longueur
tableau=[0,1,2,3,4] #création d'un tableau
print(tuple(tableau)) #affichage de la conversion de tableau en tuple
print(tuple(range(5))) #range(5) renvoie les valeurs de 0 à 4
print(tuple(4)) #message d'erreur car un entier n'est pas itérable, on ne peut pas le convertir en tuple
```

### II. Manipuler des tuples

On a déjà vu qu'on pouvait concaténer des tuples avec `+` et accèder à la valeur d'un élément seul avec `mon_tuple[i]`.

On peut également **extraire des morceaux de tuple** (slice) avec `mon_tuple[i:j]` qui renvoie un tuple composé des éléments d'indices **$i$ à $j-1$**. Assez logiquement, si **$i$ n'est pas précisé**, on extrait **à partir du premier** élément et si **$j$ n'est précisé**, on extrait **jusqu'au dernier**.

```python
#Exemple 4 :
t=(4,1,5,2,3,9,7,2,8)
print(t[2:5]) #affiche du 3ème(indice2) au 5ème(indice 4) élément
print(t[:5]) #affiche du 1er(indice 0) au 5ème(indice 4) élément
print(t[3:]) #affiche du 4ème(indice 3) au dernier élément
t2=t[:3]+(10,)+t[4:] #crée un nouveau tuple dont le 4ème élément vaut 10 au lieu de 2
print(t2) #ce nouveau tuple est une copie modifiée du premier
t[3]=10 #on ne peut pas modifier directement la 4ème valeur de t car c'est un tuple
```
