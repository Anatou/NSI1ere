## 1. Calculer une distance

En Mathématiques la distance entre deux points $A(x_A,y_A)$ et $B(x_B,y_B)$ dont on connaît les coordonnées dans un repère orthonormé est donnée par la relation : $$AB = \sqrt{(x_B - x_A)^2 + (y_B - y_A)^2}$$

**Q1. Écrire une fonction `dist_carree(A,B)` qui prend en paramètre deux tuples A et B contenant les coordonnées des points $A$ et $B$ et qui renvoie le carré de la distance AB (c'est-à-dire: AB²).**

````python
#Question 1
def dist_carree(A,B):
    '''Renvoie la distance au carrée entre les points A et B dont les coordonnées sont données sous forme de tuples (x,y,z, ...).'''
    assert (type(A) == tuple and type(B) == tuple), "A et B doivent être des tuples"
    assert len(A) == len(B), "A et B doivent être des tuples de même dimension"
    N=len(A) #nb de dimensions
    for i in range(N): #on doit vérifier que tous les éléments de A et B sont des nombres
        assert ((type(A[i])== float) or (type(A[i])== int)) and ((type(B[i]) == float) or (type(B[i])==int)), "Les valeurs de A et B doivent être des nombres"
    result = 0
    for i in range(N):
        result=result+(B[i]-A[i])**2 #on ajoute le carré de la différence pour chaque dimension
    return result
    
A=(0,0,0)
B=(2,2,2)
print(dist_carree(A,B))
````

## 2. Trouver le milieu

Les coordonnées du milieu de deux points $A(x_A,y_A)$ et $B(x_B,y_B)$ dont on connaît les coordonnées dans un repère orthonormé sont :

$$\left( \frac{x_A + x_B}{2} ,\frac{y_A + y_B}{2}\right)$$

**Q2.  Écrire une fonction `milieu(A,B)` qui prend en paramètre deux tuples A et B contenant les coordonnées des points $A$ et $B$ et qui renvoie les coordonnées de leur milieu sous forme d’un tuple.**

````python
# Question 2
def milieu(A,B):
    '''A et B sont des tuples comportant les coordonnées des points A et B. 
    Renvoie un tuple correspondant aux coordonnées (x,y,z, ...) du mileu du segment [AB].'''
    assert (type(A) == tuple and type(B) == tuple), "A et B doivent être des tuples"
    assert len(A) == len(B), "A et B doivent être des tuples de même dimension"
    N=len(A) #nb de dimensions
    for i in range(N): #on doit vérifier que tous les éléments de A et B sont des nombres
        assert ((type(A[i])== float) or (type(A[i])== int)) and ((type(B[i]) == float) or (type(B[i])==int)), "Les valeurs de A et B doivent être des nombres"
    result=() #on initialise avec un tuple vide
    for i in range(N):
        result=result+((A[i]+B[i])/2,) #pour chaque dimension on ajouteun singulet avec la moyenne des coordonnées de A et B
    return result
A=(1,2,4)
B=(3,4,4)
print(milieu(A,B))
````

````python
# Question 3
def triplet_pythagoricien(n):
    '''n est un entier >=0.
    Renvoie un tuple contenant tous les triplets pythagoriciens (i,j,k) de nombres entiers qui vérifient i^2+j^2=k^2
    avec i<=n, j<=n et k<=n.
    Renvoie un tuple vide s'il n'y en pas.'''
    assert type(n)==int and n>0, "n doit être un entier strictement positif"
    result=() #on initialise avec un tuple vide
    for i in range(1,n+1): #i varie de 1 à n
        for j in range(i,n+1): #j varie de i à n, on pourrait aussi le faire varier de 1 à i. Mais si on le fait varier de 1 à n, on obtient les doublons (i,j,k) et (j,i,k).
            for k in range(1,n+1): #k varie de 1 à n
                if k**2==(i**2+j**2) : #on teste si on a une relation de pythagore
                     result=result+((i,j,k),) # si oui, on ajoute un nouveau tuple (i,j,k)
    return result
    
print(triplet_pythagoricien(100))
print(len(triplet_pythagoricien(100)))
````
